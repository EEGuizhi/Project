### **UNet IKEM**

```python
%cd /content/Project/code

import torch
from model.unet_IKEM import UNet_IKEM
from torchsummary import summary

# Basic Settings
WITH_IGGN = True
IMAGE_SIZE = (512, 256)
NUM_OF_KEYPOINTS = 68

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = UNet_IKEM(use_iggnet=WITH_IGGN).to(device)

# Print structure & total parameters & mult-adds count
summary(
    model,
    input_data=[  # sequence of sizes
        (NUM_OF_KEYPOINTS, IMAGE_SIZE[0], IMAGE_SIZE[1]),
        (NUM_OF_KEYPOINTS, IMAGE_SIZE[0], IMAGE_SIZE[1]),
        (3, IMAGE_SIZE[0], IMAGE_SIZE[1])
    ],
    depth=7,
    col_names=["kernel_size", "input_size", "output_size", "num_params", "mult_adds"]
)
```
